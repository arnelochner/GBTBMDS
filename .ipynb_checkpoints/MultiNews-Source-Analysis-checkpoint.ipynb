```python
domains = [
    urlparse(source.split("archive.org/wayback/available?url=")[1]).netloc for source in sources
]
```

```python
domain_dict = dict()


for source in sources:
    domain = urlparse(source.split("archive.org/wayback/available?url=")[1]).netloc
    
    if domain in domain_dict.keys():
        domain_dict[domain].append(source)
    else:
        domain_dict[domain] = [source]
```

```python
def check_in_files(text):
    for i,train_file in enumerate(all_files):
        for j,ex in enumerate(train_file):
            for para in ex["src_str"]:
                if (text.lower() in para):
                    print("In Src")
                    print(f"{i} | {j}")
                    print(ex["tgt_str"])

            if (text.lower() in ex["tgt_str"]):
                    print("In TGT")
                    print(f"{i} | {j}")
                    print(ex["tgt_str"])    
```

```python
inputs_domains = []

for _input in inputs:
    url = urlparse("http://" + _input.split("//")[2].split(":")[0])
    domain = url.netloc
    path = url.path

    if "/" in path:
        path = path.split("/")[1]
        
    inputs_domains.append(domain + "/" + path)

most_common_inputs_domains = Counter(inputs_domains).most_common()[:25]
```

```python
input_domain_dict = dict()


for _input in inputs:
    domain = urlparse("http://" + _input.split("//")[2].split(":")[0]).netloc
    if domain in input_domain_dict.keys():
        input_domain_dict[domain].append(_input)
    else:
        input_domain_dict[domain] = [_input]
```

```python
input_summary_dict = dict()

for summary in summaries:
    
    summary_num = summary.split("/")[-1][:-1]
    input_summary_dict[summary_num] = {"tgt_url": summary}
                                               
                                        
for _input in inputs:
    
    
    summary_num = str(_input.split("/")[-1].split("-")[0])
    
    try:
        if "src_url" in input_summary_dict[summary_num].keys():
            input_summary_dict[summary_num]["src_url"].append(_input)
        else:
            input_summary_dict[summary_num]["src_url"] = [_input]
    except:
        print(summary_num)
```
